- 'multiple': 데이터가 겹쳐있는 상태에서 어떻게 표현할 것인가?

## Figure-level vs. axes-level functions
- Each module has a single figure-level function, which offers a unitary interface to its various axes-level functions.
- 각 모듈에는 다양한 축 수준 함수에 대한 단일 인터페이스를 제공하는 단일 Figure 수준 함수가 있습니다.
- Figuer-level functions 아래에 axes-level functions
- 즉, histplot / kdeplot... / 하위 plot 대신에 displot으로 사용 가능하다는 뜻이다


- seaborn 홈페이지
https://seaborn.pydata.org/tutorial/function_overview.html#


```python
# displot의 default값 histplot
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack");
```

```python
# displot에서 kind="kde" 사용
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack", kind="kde );
```

#### the figure-level plots(displot) 과  axes-level 의 차이점
- the legend is placed outside the plot. (displot이 범례가 밖에 있다)
- a slightly different shape
- displot이 easily create figures with multiple subplots ! ! ! 쉽게 서브플롯을 만들 수 있다 ! ! ! -> col

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species");
```

### Axes-level functions make self-contained plots
- axes-level 함수는 축 레이블과 범례를 자동으로 추가 -> legend=False로 없앨 수 있다.
- 내부적으로 matplotlib.pyplot.gca()를 호출하며(plt), 각 플롯이 어디로 가야하는지 지정하는 인수 적용 가능하다.(ax=)
- ax=axs[0]: scatterplot을 첫 번째 틀에 넣는다
- shrink=.8: 막대 그래프 사이 간격
- alpha=.8: 막대 색 진하기

```python
# 그래프 틀 만들기 
f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", ax=axs[0])
sns.histplot(data=penguins, x="species", hue="species", shrink=.8, alpha=.8, legend=False, ax=axs[1]);
```

### Figure-level functions own their figure
- In contrast, figure-level functions cannot (easily) be composed with other plots.
- relplot: 산점도 plot
- xy1=(10, 2), slope=.2: 기울기 정도와 기울기 그래프에 대한 위치
- dashes=(5,2): 점선 표현 길이와 점선의 점 사이 간격

```python
tips = sns.load_dataset("tips")
g = sns.relplot(data=tips, x="total_bill", y="tip")
g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(5,2))
```

### Customizing plots from a figure-level function

```python
g = sns.relplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", col="sex")
g.set_axis_labels("Flipper length(mm)", "Bill length (mm)");

# only when using a figure-level function.
```

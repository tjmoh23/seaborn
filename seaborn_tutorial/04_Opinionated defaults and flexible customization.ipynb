```python
%matplotlib inline
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
penguins.head(2)
```

```python
sns.relplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g");
```

참고 문헌: https://greeksharifa.github.io/machine_learning/2019/12/05/Seaborn-Module/

- sns.set_theme(style=None, rc=None): 그래프 배경 설정
    - style = “darkgrid”, “whitegrid”, “dark”, “white”, “ticks”
    - rc = [dict], 세부 사항을 조정함
- sns.despine(offset=None, trim=False, top=True, right=True, left=False, bottom=False)
    - top, right, left, bottom = True로 설정하면 그 축을 제거함
    - offset = [integer or dict], 축과 실제 그래프가 얼마나 떨어져 있을지 설정함

```python
sns.set_theme(style="ticks", font_scale=1.25 )
g = sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g",
    palette="crest", marker="x", s=100,
)
g.set_axis_labels("Bill length (mm)", "Bill depth (mm)", labelpad=10)
g.legend.set_title("Body mass (g)")
g.figure.set_size_inches(6.5, 4.5)
g.ax.margins(.15)
g.despine(trim=False);

# style="ticks": 그래프 뒷배경 흰색
# s=100: 점 크기
# labelpad: 축 제목과 그래프 사이 간격
# ax.margins(.15) 점의 분포도
```
